```python
from PyPDF2 import PdfFileReader, PdfFileWriter
```

```python
pdf_reader = PdfFileReader(open('aa2.pdf', 'rb'))
```

```python
if pdf_reader.isEncrypted:
    pdf_reader.decrypt('')
#else:
    # PDF is not encrypted, no need to remove password restrictions
```

```python
pdf_writer = PdfFileWriter()

if not pdf_reader.isEncrypted:
    # If the PDF was not encrypted, add all pages directly
    for page_num in range(pdf_reader.numPages):
        pdf_writer.addPage(pdf_reader.getPage(page_num))
else:
    # If the PDF was encrypted and the password restrictions were removed, add all pages
    for page_num in range(pdf_reader.numPages):
        page = pdf_reader.getPage(page_num)
        pdf_writer.addPage(page)
```

```python
with open('aa23.pdf', 'wb') as output_pdf:
    pdf_writer.write(output_pdf)
```
