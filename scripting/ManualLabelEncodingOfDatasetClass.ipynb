# AFB

```python
afb =  read_excel('./datasets/New_datas_first.xlsx')
target = 'ENIMPAYEOUPAS'
afb.head()
```

```python
# get unique values
uniq_afb_class = list(afb[target].unique())
uniq_afb_class
```

```python
# replacement
# IMPAYE == 1
# PAYE == 0
replacements = {'IMPAYE': 1, 'PAYE': 0}
afb[target] = afb[target].replace(replacements)
print(afb.head())
print(afb.describe())
```

```python
# check the type of class attribut
print(afb.info())
```

```python
# save the actual version to source
afb.to_excel('./datasets/New_datas_first.xlsx')
```

# GERMAN

```python
germ = read_csv('./datasets/private/1. UCI Repository/German/german.csv', sep=',')
target = 'Class'
germ.head()
```

```python
uniq_germ_class = list(germ[target].unique())
uniq_germ_class
```

```python
# replacement based on signification of actual values
# 1 means Good Loans and 
# 2 means Bad loans
germ[target] = germ[target].replace(1,0)
germ[target] = germ[target].replace(2,1)
germ.describe()
```

```python
uniq_germ_class = list(germ[target].unique())
uniq_germ_class
```

```python
germ.to_csv('./datasets/private/1. UCI Repository/German/german.csv', sep=',', index=False)
```
