```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Charger les données
df = pd.read_csv('dataS.csv')

# Exercice 1 : Catégorie des variables
# (Déjà fait dans la réponse précédente, c'est une tâche conceptuelle)

# Exercice 2 : Tableau statistique
def tableau_statistique(df):
    stats = {
        'Sexe': df['sexe'].value_counts().to_dict(),
        'Notes': {
            'Moyenne': df['Notes'].mean(),
            'Min': df['Notes'].min(),
            'Max': df['Notes'].max()
        },
        'Age': {
            'Moyenne': df['Age'].mean(),
            'Min': df['Age'].min(),
            'Max': df['Age'].max()
        },
        'Taille': {
            'Moyenne': df['Taille'].mean(),
            'Min': df['Taille'].min(),
            'Max': df['Taille'].max()
        },
        'Couleur': df['Couleur'].value_counts().to_dict()
    }
    return stats

# Exercice 3 : Diagrammes
def diagramme_couleur(df):
    plt.figure(figsize=(10, 6))
    df['Couleur'].value_counts().plot(kind='pie', autopct='%1.1f%%')
    plt.title('Répartition des couleurs')
    plt.ylabel('')
    plt.show()

def diagramme_taille(df):
    plt.figure(figsize=(10, 6))
    sns.histplot(df['Taille'], kde=True)
    plt.title('Distribution des tailles')
    plt.xlabel('Taille (m)')
    plt.ylabel('Fréquence')
    plt.show()

# Exercice 4 : Paramètres de position
def parametres_position(df, variable):
    return {
        'Mode': df[variable].mode().values[0],
        'Moyenne': df[variable].mean(),
        'Médiane': df[variable].median()
    }

# Exercice 5 : Paramètres de dispersion
def parametres_dispersion(df, variable):
    return {
        'Étendue': df[variable].max() - df[variable].min(),
        'Variance': df[variable].var(),
        'Écart-type': df[variable].std()
    }
```

```python

# Exécution
print("Exercice 2 :")
tableau_statistique(df)

```

```python
print("\nExercice 3 :")
diagramme_couleur(df)
diagramme_taille(df)
```

```python

print("\nExercice 4 :")
print("Notes :", parametres_position(df, 'Notes'))
print("Âge :", parametres_position(df, 'Age'))

```

```python
print("\nExercice 5 :")
print("Notes :", parametres_dispersion(df, 'Notes'))
print("Âge :", parametres_dispersion(df, 'Age'))
```
