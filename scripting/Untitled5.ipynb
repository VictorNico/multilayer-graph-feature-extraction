```python
content = """<?xml version="1.0" encoding="UTF-8"?>
<xsl:stylesheet version="1.0"
xmlns:xsl="http://www.w3.org/1999/XSL/Transform">
  <xsl:template match="/">
    <html>
      <body>
        <h2>Contenu XML transformé en HTML</h2>
        <ul>
          <xsl:for-each select="items/item">
            <li><xsl:value-of select="name"/></li>
          </xsl:for-each>
        </ul>
      </body>
    </html>
  </xsl:template>
</xsl:stylesheet>
"""
with open("fichier.xslt", "w") as output:
    output.write(str(content))
```

```python
from lxml import etree

# Chargement du XML et XSLT
xml_file = etree.parse("output.xml")
xslt_file = etree.parse("fichier.xslt")
transform = etree.XSLT(xslt_file)

# Transformation
result = transform(xml_file)

# Sauvegarde en HTML
with open("result.html", "w") as output:
    output.write(str(result))
```
