```python
from PIL import Image

# Charger les images à coller
image1 = Image.open('AFB_dtc_shapley.png')
if image1.mode != 'RGBA':
    print('###')
    image1 = image1.convert('RGBA')
image2 = Image.open('CREDIT_RISK_DATASET_dtc_shapley.png')
if image2.mode != 'RGBA':
    print('###')
    image2 = image2.convert('RGBA')
image3 = Image.open('GERMAN_dtc_shapley.png')
if image3.mode != 'RGBA':
    print('###')
    image3 = image3.convert('RGBA')
image4 = Image.open('JAPAN_dtc_shapley.png')
if image4.mode != 'RGBA':
    print('###')
    image4 = image4.convert('RGBA')

# Redimensionner les images pour qu'elles s'adaptent à une grille 2x2 sur une page A4
page_width, page_height = 2480, 3508  # Dimensions en pixels pour une page A4 (300 dpi)
grid_cols, grid_rows = 2, 2
image_width = page_width // grid_cols
image_height = page_height // grid_rows

image1 = image1.resize((image_width, image_height))
image2 = image2.resize((image_width, image_height))
image3 = image3.resize((image_width, image_height))
image4 = image4.resize((image_width, image_height))

# Créer une image vide au format A4 pour la grille
grid_image = Image.new('RGBA', (page_width, page_height), (0, 0, 0, 0))

# Coller les images dans la grille
grid_image.paste(image1, (0, 0), mask=image1)
grid_image.paste(image2, (image_width, 0), mask=image2)
grid_image.paste(image3, (0, image_height), mask=image3)
grid_image.paste(image4, (image_width, image_height), mask=image4)

# Sauvegarder l'image de la grille au format A4
grid_image.save('dtc_shapley.png')

# Afficher l'image de la grille
grid_image.show()
```
