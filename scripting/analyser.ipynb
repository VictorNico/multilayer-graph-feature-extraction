```python
import os

# Specify the root directory
root_dir = './outputs/mlna_1/data_selection_storage/CREDIT_RISK_DATASET/'

def get_importance_data_filenames(root_dir, func, verbose=False):
    data_filenames = []
    # Walk through the directories and files
    for dirpath, dirnames, filenames in os.walk(root_dir):
        # dirpath: current directory path
        # dirnames: list of directories in the current directory
        # filenames: list of files in the current directory

        # Print the current directory
        print('Directory:', dirpath)  if verbose else None
        # Print all the subdirectories
        for dirname in dirnames:
            print('Subdirectory:', os.path.join(dirpath, dirname))  if verbose else None

        # Print all the files
        for filename in filenames:
            if func(filename) and not ('x_' in filename or 'y_' in filename or 'metric' in filename):
                print('File:', os.path.join(dirpath, filename)) if verbose else None
                data_filenames.append(os.path.join(dirpath, filename))

        # Print an empty line to separate directories
        print()  if verbose else None
    return data_filenames
```

```python
def get_mlna_1_degree_details(root_dir, model, func, func_2):
    filenames = get_importance_data_filenames(root_dir,func) # get list of degree ones
    # define outputs columns
    cols = [
    'Rang',
    'Coef. importance',
    'Meil coef. importance',
    'Accuracy',
    'Précision',
    'Rappel',
    'F1-mesure'

    ]
    _store = pd.DataFrame(columns=cols)
    for file in filenames: # fetch it
        dataset = load_data_set_from_url(path=file,sep='\t', encoding='utf-8',index_col=0) # load the file
        # get model information
        target = dataset.copy()
        # get metrics
        #print(target.index.values.tolist())
        metrics = target.loc[model,['accuracy', 'precision', 'recall', 'f1-score']].values
        #print(metrics)
        # drop metrics column in dataframe
        target.drop(['precision', 'accuracy', 'recall', 'f1-score'], axis=1, inplace=True)
        # sorting
        _sorted = target.sort_values(
            by = model,
            axis = 1, 
            ascending = False
            )
        # get degree variable
        degree = [(index+1,var) for index,var in enumerate(_sorted.columns.to_list()) if func_2(var)][0]
        # ajoutons le descriptuer dans le store
        #print(_sorted.loc[model,[degree[1]]].values[0])
        #print(_sorted.loc[model,:].values[len(_sorted.columns.to_list())-1])
        vals = [
        str(degree[0])+'/'+str(len(_sorted.columns.to_list())), # the rang
        _sorted.loc[model,[degree[1]]].values[0], # the coeficient importance
        _sorted.loc[model,:].values[0] # the best coeficient importance
        ]
        #print(vals)
        vals.extend(metrics)

        _store.loc[degree[1]] = pd.Series(
            vals, 
            index=cols
            #,name= degree[1]
            )
    return _store
```

```python
mlna_1_mln = {
    model : get_mlna_1_degree_details(
        root_dir = root_dir, 
        model = model, 
        func = lambda x: (('classic_mln_' in x) and not('classic_mln_-' in x)),
        func_2 = lambda x: ('degree' in x)
    ) for model in models
}
```
